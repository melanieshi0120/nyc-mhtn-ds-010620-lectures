```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy.random import seed, randint
seed(100)

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=True, dpi=100)

# Define the colors and markers to use
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Plot each axes
for i, ax in enumerate(axes.ravel()):# list four graphs(arrary type) into list type  
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('left')

plt.suptitle('Four Subplots in One Figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
plt.show()
```

```python
from numpy.random import seed, randint
seed(100)

x = sorted(randint(0,10,10))
x2 = sorted(randint(0,20,10))
y = sorted(randint(0,10,10))
y2 = sorted(randint(0,20,10))
```

- Plot 1: a line blue graph (`.plot()`) using data `x` and `y`

```python
fig, axes = plt.subplots(2,2, figsize=(10,6), dpi=100)
axes[0][0].plot(x,y)
axes[0][0].set_title('Title1',color='b')

axes[0][1].scatter(x2,y2,color='r')
axes[0][1].set_title('Title2',color='b')

axes[1][0].plot(x,y)
axes[1][0].scatter(x2,y2,color='r')
axes[1][0].set_title('Title3',color='b')

axes[1][1].plot(x,y)
axes[1][1].scatter(x2,y2,color='r')
axes[1][1].yaxis.set_ticks_position('right')
axes[1][1].set_title('Title4',color='b')

```

```python
# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(8,6), sharex=False, sharey=False, dpi=120)
axes[0][0].plot(x,y, color='b' )
axes[0][0].set_title('Ax: 1')

axes[0][1].scatter(x2,y2, color='r', facecolor='none' )
axes[0][1].set_title('Ax: 2' )

axes[1][0].plot(x,y, color='b' )
axes[1][0].scatter(x2,y2, color='r', marker='o', facecolor='none' )
axes[1][0].set_title('Ax: 3' )

axes[1][1].plot(x,y, color='b' )
ax2 = axes[1][1].twinx()# share one x axis and using two differet y-axis with different scales
#ax2 = axes[1][1].twiny()#
ax2.scatter(x2,y2, color='r', marker='o', facecolor='none' )
axes[1][1].set_title('Ax: 4' )
fig.suptitle('Four Subplots in One Figure', fontsize=16)  
fig.savefig('answer_guide')
```
